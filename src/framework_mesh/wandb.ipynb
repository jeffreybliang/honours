```python
import wandb
```

```python
api = wandb.Api()
```

```python
run = api.run("u7013004-the-australian-national-university/Test/0rid6g2s")
```

```python
history = run.scan_history(keys=["plt/heatmap"], min_step=100)
```

```python
import plotly.io as pio

for row in history:
    if row.get("plt/heatmap") is not None:
        heatmap_ref = row["plt/heatmap"]
        break  

file = run.file(heatmap_ref['path'])
local_path = file.download(replace=True).name  # downloads and gives local filename

# Now load the Plotly figure from the downloaded JSON file
fig = pio.read_json(local_path)

# Show or save
fig.show()
```

```python
fig = pio.read_json(local_path)

# --- REMOVE AXES ---
fig.update_layout(
    xaxis=dict(visible=False),
    yaxis=dict(visible=False),
    plot_bgcolor='rgba(0,0,0,0)',   # transparent plot background
    paper_bgcolor='rgba(0,0,0,0)'   # transparent full figure background
)
fig.update_layout(
    scene = dict(
        xaxis = dict(visible=False),
        yaxis = dict(visible=False),
        zaxis =dict(visible=False)
        )
    )
fig.update_layout(
    margin=dict(l=0, r=0, t=0, b=0)
)

fig.data[0].update(
    cmin=-.1,
    cmax=.1,
    colorscale="Viridis",  # or any other name, or a custom list,
    showscale=False
)
fig.show()
```
